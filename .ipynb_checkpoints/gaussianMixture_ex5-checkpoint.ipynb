```python
#prova del gaussianmixture sui dati considerano N = 1, 2, .., 10 gaussiane: la migliore scelta è quella che minimizza l'AIC
gm_N = []
N = np.linspace(1,10,10, dtype=int)
for n in N:
    gm_N.append(mix.GaussianMixture(n).fit(data).aic(data))
    
plt.plot(N,gm_N, '-o')
```

```python
#fit dei dati con la migliore scelta trovata (minima AIC)
i_min = gm_N.index(np.min(gm_N))
gm_min = mix.GaussianMixture(N[i_min]).fit(data)
y_min = gm_min.score_samples(np.linspace(np.min(data),np.max(data),101)[:,np.newaxis])

plt.hist(data, bins=np.linspace(np.min(data),np.max(data),101), density=True)
plt.plot(np.linspace(np.min(data),np.max(data),101), np.exp(y))
```
